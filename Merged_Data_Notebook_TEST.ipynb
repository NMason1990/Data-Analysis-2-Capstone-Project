```python
import pandas as pd
```

```python
scraped_data = pd.read_csv('fab_scraped_data.csv')

scraped_data.rename(columns={'rarity':'rarity', 'product':'name'}, inplace = True)
scraped_data.head(15)
```

```python
api_data = pd.read_csv('API_data.csv')
api_data.drop('stats', axis=1, inplace=True)
api_data.head(2)
```

```python
import pandas as pd

#Changing the column's below to numeric data type, removing the decima, and filling NaN values with 0

columns_to_convert = ['pitch value', 'life', 'intellect', 'attack', 'defense', 'cost']

# Fill NaN values with 0 and then convert to integer data type
api_data[columns_to_convert] = api_data[columns_to_convert].fillna(0).astype(int)

# Print the DataFrame to verify the changes
api_data.head(15)
```

```python
# Perform the join based on the remaining 'rarity' colum   
#merged_data = api_data.merge(scraped_data, on='rarity', how='left')
merged_data = pd.merge(api_data, scraped_data, on=['number'], how='left')
merged_data.head(10)
```

```python
merged_data.head(2)
```

```python
import pandas as pd
# creating a variable to loop the two columns and clean the data
columns_to_clean = ['market price', 'listed median']
# for loop of both columns in dataframe
for column in columns_to_clean:
    merged_data[column] = merged_data[column].replace({'\$': ''}, regex=True)  # Remove "$" sign
    merged_data[column] = pd.to_numeric(merged_data[column], errors='coerce').fillna(0.00)  # Convert to numeric and fill NaN with 0.00
    
print(merged_data)
```
